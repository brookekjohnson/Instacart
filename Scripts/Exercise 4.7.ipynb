# 03. Creating New Variables

```python
#create subset
df = df_ords_prods[:1000000]
```

```python
df.shape
```

```python
df.head()
```

```python
#define variable for price_label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply price_label to subset to create new column 'price_range'
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
#create new column using different method
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#frequencies of orders on each day of the week
df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

```python
# create for loop to indicate level of busy-ness of any given day
result = []

for value in df_ords_prods["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create a new column with for loop
df_ords_prods['busiest_day'] = result
```

```python
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

```python
#create new for loop on busiest days to accomodate clients
result = []

for value in df_ords_prods["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Least busy days")
  else:
    result.append("Regularly busy")
```

```python
#create new column for busiest days
df_ords_prods['busiest_days'] = result
```

```python
#check values for busiest_days column
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

# The values check makes sense because the busiest days is within range of twice that of the busiest day that we creared previously, which is accuate based on the values check for the previous for loop.  The same logic applies for the least busy days.

```python
#value check for order_hour_of_day
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

```python
#create for loop for busiest_period_of_day
result = []

for value in df_ords_prods["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    result.append("Most Orders")
  elif value == 7 or value == 8 or (value >= 17 and value <= 22):
    result.append("Average Orders")
  else:
    result.append("Fewest Orders")
```

```python
#create new column for busiest_period_of_day
df_ords_prods['busiest_period_of_day'] = result
```

```python
# check frequency for new column
df_ords_prods['busiest_period_of_day'].value_counts(dropna =False)
```
