# 05. Excluding Low-activity Customers

```python
#check variable names
df.columns
```

```python
#create exclusion flag for low activity customers (less than 5 orders)
def exclusion_flag(row):
    if row['max_order'] < 5:
        return 'Yes'
    else: return 'No'
```

```python
#check data
df.head()
```

```python
#check shape
df.shape
```

```python
#apply exclusion flag
df['exclude'] = df.apply(exclusion_flag, axis = 1)
```

```python
#check exclusion
df['exclude'].value_counts(dropna = False)
```

```python
#check data
df.head()
```

```python
#check shape
df.shape
```

```python
#drop exclusions
df.drop(df.loc[df['exclude'] == 'Yes'].index, inplace = True)
```
