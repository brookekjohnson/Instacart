# 03. Creating Groups

```python
# calculate aggregated mean of order_number grouped by department_id
df.groupby('department_id')['order_number'].mean()
```

# the means of the entire dataframe are close to, but not the exact same as those in the subset, which makes sense because there are a lot more numbers being calculated into the mean of the entire dataframe.

# loyalty flag was created in previous exercise and can be seen in the dataframe head function above

```python
#compare basic statistics of different types of customers
df.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#calculating average prices for each user
df.groupby('user_id').agg({'prices': ['mean']})
```

```python
#create new column for average prices
df['average_prices'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df.head()
```

```python
df.shape
```

```python
# create spending_flag
df.loc[df['average_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df.loc[df['average_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df.head(100)
```

```python
#creating new column for median of days_since_prior_order
df['median_days_since_prior_order'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df.head()
```

```python
#creating frequency flag
df.loc[df['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['median_days_since_prior_order'] > 10) & (df['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df.loc[df['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
