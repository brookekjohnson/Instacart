# 03. Creating Visualizations

```python
#create histogram for order_hour_of_day
df['order_hour_of_day'].plot.hist(bins = 30)
plt.title('Frequency of Orders by Time of Day')
plt.xlabel('Hour of Day')
plt.savefig(r'C:\Users\Brooke\05-2022 Instacart Basket Analysis\Analysis\Visualizations\time of day hist.png')
```

# This histogram demonstrates that sales spike in the middle of the day, beginning a little bfore 10, declining a little after 5, and peaking around 10.  It also forms the shape of a normal distribution.

```python
# create a bar chart for loyalty_flag
df['loyalty_flag'].value_counts().plot.bar()
```

```python
# create random sample
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
#store 70% of the sample in dataframe big
big = df[dev]
```

```python
#store 30% of the sample in dataframe small
small = df[~dev]
```

```python
#check sample dataframes
len(df)
```

```python
len(big) + len(small)
```

```python
#assign small dataframe columns prices and hour of day to dataframe df_2
df_2 = small[['order_hour_of_day','prices']]
```

```python
#create line chart
line = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

```python
#assign small dataframe columns age and number of dependants to dataframe df_3
df_3 = small[['Age', 'number_of_dependants']]
```

```python
#create line chart 
line_2 = sns.lineplot(data = df_3, x = 'Age',y = 'number_of_dependants')
```

# This line chart shows that the relationship between age and number of dependents is fairly random.  There is not an obvious trend in the data which is surprising, as I would have suspected the number of dependents to be higher among the young to middle aged adults than the elder adults.

```python
#create a scatter plot
sns.scatterplot(x = 'Age', y = 'income',data = df)
```

# This scatterplot shows that the mass majority of people have incomes below 200,000, but it is clear also that the majority of people who do make over 200000 are above the age of 40, which means that age group would have more spending power in general than the younger age group.
